```python
from pycocotools.coco import COCO
import numpy as np
import skimage.io as io
import matplotlib.pyplot as plt
%matplotlib inline

json_path = 'C:\\ANN\\Dataset\\COCO\\annotations\\instances_train2017.json'
coco = COCO(json_path)

catIds = coco.getCatIds(catNms=['dog'])
imgIds = coco.getImgIds(catIds=catIds)
img = coco.loadImgs(imgIds[np.random.randint(0,len(imgIds))])[0]

I = io.imread('C:\\ANN\\Dataset\\COCO\\train2017\\' + img['file_name'])
plt.imshow(I)
plt.show()
```

```python
plt.imshow(I)
annIds = coco.getAnnIds(imgIds=img['id'])
anns = coco.loadAnns(annIds)
coco.showAnns(anns)
plt.show()
```

```python
import os
import numpy as np
from pycocotools.coco import COCO
import matplotlib.pyplot as plt
from PIL import Image
os.chdir('C:\\ANN\\Dataset\\COCO\\trainLabels2017')

json_path = 'C:\\ANN\\Dataset\\COCO\\annotations\\instances_train2017.json'
coco = COCO(json_path)

imgIDs = coco.getImgIds()
imgs = coco.loadImgs(imgIDs)

for img in imgs:
    anno_ids = coco.getAnnIds(img['id'])
    anns = coco.loadAnns(anno_ids)

    mask = np.zeros((img['height'], img['width']), dtype=np.int32)
    for i in range(len(anns)):
        mask = np.maximum(coco.annToMask(anns[i]) * anns[i]['category_id'], mask)
        pil_msk = Image.fromarray(mask)
        pil_msk.save(img['file_name'].replace('.jpg','.png'))

print('done')
```
